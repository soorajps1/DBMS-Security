```python
import mysql.connector
from sqlalchemy import create_engine
import pandas as pd

# Replace 'your_username', 'your_password', 'your_database', and 'your_host' with actual values
db_connection = mysql.connector.connect(
    user='root',
    password='12345',
    host='localhost',
    database='bat_algo_database'
)

# Path to your CSV file
csv_file_path = r'C:\Users\user\BAT_ALGORITHM_FULL\processed_data_bat_full.csv'

# Read CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path)

# Define the table name (avoid spaces)
table_name = 'data_collected'

# Create SQLAlchemy engine
engine = create_engine('mysql+mysqlconnector://root:12345@localhost/bat_algo_database')

try:
    # Create MySQL table based on DataFrame structure
    df[:0].to_sql(table_name, con=engine, index=False, if_exists='replace')

    # Load data into MySQL table
    df.to_sql(table_name, con=engine, index=False, if_exists='append')

    print(f'Data has been successfully loaded into the {table_name} table.')

except Exception as e:
    print(f'Error: {str(e)}')

finally:
    # Close the database connection
    db_connection.close()
```
